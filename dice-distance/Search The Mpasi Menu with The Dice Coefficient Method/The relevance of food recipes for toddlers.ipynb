## Import libraries

```python
from dice_distance import DiceDistance
import pandas as pd
```

## Using Custom Libraries

```python
dataset = pd.read_excel('datasets.xlsx')
diceDistance = DiceDistance(dataset=dataset)
```

```python
query = "daging ayam makaroni tahu"
dataset['Scores'] = diceDistance.search(query=query)
dataset['Information'] = diceDistance.relevantChecker()
```
